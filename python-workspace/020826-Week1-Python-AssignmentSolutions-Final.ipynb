```python
#IK Python Practice Set: Week 1
#Problem #4 - #Intersection Of Two Arrays
#Given two lists of numbers, find their intersection.
#a = [4, 2, 2, 3, 1]
#b = [2, 2, 2, 3, 3]

a = [1000000, 1000000, 0, 1, -13, 5, 0]
b = [0, 5, -13, 1, 0, 1000000, 1000000]

#a = [1]
#b = [1]


result = list()

#Solution 1: Using two for loop and removing matched element
#This is not perfect solution as it is exceeding the time limit for large arrays
'''
for x in a:
    if x in b:
        result.append(x)
        b.remove(x)
'''

#Solution 2: Using sorting and comparing two arrays in a while loop

a_sorted = sorted(a)
b_sorted = sorted(b)
i = 0
j = 0

while i < len(a_sorted) and j <= len(b_sorted) :

    if a_sorted[i] == b_sorted[j] :
        #print("match found:", a_sorted[i])
        result.append(a_sorted[i])
        i += 1
        j += 1
        continue

    elif a_sorted[i] < b_sorted[j] :
        i += 1

    else :
        j += 1

print("Matching numbers:",result)
```

```python
#IK Python Practice Set: Week 1
#Problem #6 - #Reverse the given string

s = "best technical interview prep courses"

word_list = s.split()

#Solution 1: Using Built-in function
#r_word_list = " ".join(reversed(word_list))

#Solution 2: Brute Force
r_word_list = []
n = len(word_list)

for i in range(n-1, -1, -1):
    r_word_list.append(word_list[i])

r_words = " ".join(r_word_list)

print(r_words)
```

```python
#IK Python Practice Set: Week 1
#Problem #7 - #First Occurrence Of A Character
#Find the first occurrence of a given character in a given string

s = "b"
to_find = "c"
ind = -1
count = 0
#Solution 1 : Using Built-in function
#ind = s.find(to_find)

#Solution 2 : Using list, for loop
clist = list(s)

for c in clist :
    if c == to_find:
        ind = count
        break
    count += 1

print(ind)
```

```python
#IK Python Practice Set: Week 1
#Problem #8 - #Count Alphabets
#Count the number of alphabets in a given string.
s="abcd123BA"

count = 0
clist = list(s)

#Solution 1 : Using Built-in function
for c in clist :
    if c.isalpha():
        count += 1


#Solution 2 : unicode/ascii value check using ord function
scnt = 0
ccnt = 0
aord = ord('a')
zord = ord('z')
Aord = ord('A')
Zord = ord('Z')

for c in clist :
    #if (ord('a') <= ord(c) <= ord('z')) or (ord('A') <= ord(c) <= ord('Z')) :
    if aord <= ord(c) <= zord:
        scnt += 1
    if Aord <= ord(c) <= Zord:
        ccnt += 1

count = scnt+ccnt

print(count)
```

```python
#IK Python Practice Set: Week 1
#Problem #9 - #Convert Uppercase Characters To Lowercase
#Convert all uppercase characters to lowercase in a given string.

s = "KICKSTART INTERVIEw     ML/AI 12ABC 12 S"
clist= list(s)
nlist = list()
A_ord = ord('A')
Z_ord = ord('Z')
newstr = ""

#Solution 1: Built-in function
#print(s.lower())

#Solution 2: Using list and for loop
for c in clist:
    if A_ord <= ord(c) <= Z_ord:
        nlist.append(c.lower())
    else:
        nlist.append(c)

newstr = "".join(nlist)

#Solution 3: Using list comprehension
#newstr = [ c.lower() for c in clist]  #No condition
newstr = [ c.lower() if A_ord <= ord(c) <= Z_ord else c for c in clist] #list comprehension with ascii value check using ord


#Solution 4: Using lambda function
newstr = list(map(lambda c: c.lower(), clist))
newstr = "".join(newstr)

print(newstr)
```
