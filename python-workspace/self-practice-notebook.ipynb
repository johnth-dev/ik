```python
a = [7, 7, 14, 92, 14, 92, 92]
b = [0, 0, 92, 92, 7]

intersection = list()
for ia in a:
    if ia in b:
        intersection.append(ia)
        b.remove(ia)
print(intersection)

```

```python
nums = [2,3,5]
rval = -1
for i in range(len(nums)):
    #print("index:", i)
    right = nums[i+1:]
    left = nums[0:i]
    if sum(left) == sum(right):
        #print("sum matched at", i)
        #print(left,":", right)
        rval = i
        break
print(rval)
```
