#Multiple inheritance

```python
class A:
    def method_a(self):
        print("method_a from class A")

    def shared_method(self):
        print("Shared method A")

class B:
    def method_b(self):
        print("method_b from class B")

    def shared_method(self):
        print("Shared method B")

class C(A, B): #C is child of A and child of B - multiple inheritance
    def method_c(self):
        print("method_c from class C")

objC = C()
objC.method_a()
objC.method_b()
objC.method_c()
objC.shared_method() #both A & B both has shared method. interpreter picks the shared_method() from A. read more about this.

objA = A()
objA.method_c() #A is parent of C. A will not have access to Object C methods. This will give error


### READ ABOUT PRIVATE AND PROTECTED VARIABLES AND METHODS
```

```python
#Sorting String Arrays
words = ["banana","date","fig","cherry","bananas","apple"]
print(sorted(words)) # sorted function doesn't change the array itself. original order will be retained
print(words) #['banana', 'date', 'fig', 'cherry', 'bananas', 'apple']
words.sort() # this change the original array
print(words) #['apple', 'banana', 'bananas', 'cherry', 'date', 'fig']

#sorting by length
def getLength(word):
    return(len(word))

#Sorting based on word length using a user defined function. #sorted the words array based on the length using the function defined
print("Sorting using user defined function:",sorted(words, key=getLength))

#Same using lambda
print("Using lambda function:", sorted(words, key=lambda x: len(x)))

#Even simpler is
print("Using sleeker function:", sorted(words,key=len))

#Sorting based on the first letter of the each words
print("Sorted using the second letter of the words:", sorted(words,key=lambda x: x[1]))

#Sorting based on the first & second letter of the each words
print("Sorted using the second and first letter of the words:", sorted(words,key=lambda x: (x[1], x[0])))

```

```python
#TypeError: Converting a non-numeric to string
try:
    b = "abc" + 3
    x += 2
    x /= 0
except:
    print("Cannot add int to a string")

print("reaches this line even though above expression was erroneous")

# Improved Exception handling
try:
    # c = int("string")
    # b = "abc" + 3
    x = 1
    x += 2
    x /= 0
except Exception as ex:
    print("Oops!.. something went wrong")
    print(type(ex))
    print(ex)
    parent_classes = ex.__class__.__bases__
    for parent in parent_classes:
        print(f"Direct parent class:{parent.__name__}")
        grandparent_classes = parent.__bases__
        for grandparent in grandparent_classes:
            print(f"Grandparent class:{grandparent.__name__}")
```

```python
## Raise an Error
def divide(a,b):
    if b==0:
        manual_exception = Exception("Cannot divide by zero - raising a custom exception")
        print(f"Exception has type {type(manual_exception)}")
        raise manual_exception
    return a/b

try:
    divide(12,0)
except Exception as ex:
    print(ex)
```

Custom Exception - creation

```python
#Custom Exception

#Create Custom Exception

class CustomError(Exception): #new CustomError class created inherited from Exception base class
    def __init__(self, message):
        self.message = message


try:
    raise CustomError("This is a custom error message raised")
except Exception as ex:
    print(type(ex))
    print(ex.message)

```

JSON Module

```python
#Serialize to JSON string
#JSON is a pure string with a dictionary structure
person = {
    "name": "Tom",
    "age": 50,
    "city": "Fairfield",
    "hasChildren": True,
    "titles": ["engineer", "programmer"]
}

#Converting the above String json object to Json object
import json
person_json = json.dumps(person)

print(person_json)

## Formatting JSON object -- JSON.DUMPS
person_json = json.dumps(person, indent=7, separators=(",",": "), sort_keys=False)
print(person_json)


## Formatting JSON object -- JSON.DUMP ***** different from DUMPS
## DUMP writes the json object into a file
with open("json_dump_sample.json","w") as file:
    json.dump(person,file)
```
