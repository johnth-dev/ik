```python
#IK Week 2:Squares Of A Sorted Array
#Problem #5: Given an array of numbers sorted in increasing order, generate another array containing the square of all the elements in the given array, sorted in increasing order.

#approach 1:
#square, append, and sort
# lambda - this is sorting.. time : nlog n, space: constant

#approach 2:
#use 2 pointer approach and compare the squire of the corner values and append the greater values in the result
# time: N, space constant. THis is more efficient.

n=len(arr)
res=[0]*n # this is initializing the size of the array with size n.

l=0
r=n-1
idx=n-1
while l<r:
    if arr[l]*arr[l] <= arr[r]*arr[r]
        res[idx] = arr[r]*arr[r]
        r=r-1
        idx=idx-1
    else:
        res[idx] = arr[l]*arr[l]
        l=l+1
        idx=idx-1
    return res
```
