Handling Missing Values

```python

import numpy as np
student_data = {
    'account_length': [20,20,np.nan, 20,30],
    'gender': [np.nan, 'M', 'M', 'F', 'F'],
    'education': ['high school', 'high school', 'high school', 'high school', np.nan]
}
df = pd.DataFrame(student_data, index=['a','b','c','d','e'])
df.reset_index()

#Replace the missing value with average value of the column 'account_length'
value_to_replace = df['account_length'].mean()
# print(value_to_replace)

# - this action will not affect the original dataframe. only on the copy of the dataframe.
# df['account_length'].fillna(value_to_replace)

# - this action will not affect the original dataframe. only on the copy of the dataframe.
# df['account_length'].fillna(value_to_replace, inplace=True)

## This is the better way of doing the fillna. direct method on the dataframe
# df.fillna({'account_length': value_to_replace}, inplace=True)
# df.fillna({'gender':'M'}, inplace=True)

## Mode 0 is required to get the value as mode() returns a pd.Series
mode_value = df['education'].mode()[0]

# print(type(mode_value))
# df.fillna({'education':mode_value}, inplace=True)

# Submitting the fillna in one statement for all columns
df.fillna(
    {
        'account_length': value_to_replace,
        'education': mode_value,
        'gender': 'F'
    }, inplace=True
)

```

Sorting in a dataframe

```python
product_data = {
    'Date': ['2023-09-01','2023-09-02','2023-09-03','2023-09-04'],
    'Product': ['Product A','Product B','Product A','Product C'],
    'Quantity Sold': [100, 150, 120, 80],
    'Revenue': [5000,7500,6000,4000]
}
df = pd.DataFrame(product_data)
# df.sort_values(by='Revenue', ascending=False)

#reset_index preserves the old and new index
df.sort_values(by='Revenue', ascending=False).reset_index()

#drop the old index and retains the new index after the sorting
df.sort_values(by='Revenue', ascending=False).reset_index(drop=True)
```

```python
student_score_data = {
    'Student_Name': ['Alice', 'Bob', 'Eve', 'David', 'Carol'],
    'Age': [25,30,22,24,28],
    'Score': [95,93,95,88,88]
}
df = pd.DataFrame(student_score_data)
df.sort_values(by='Score', ascending = False).reset_index(drop=True)

# Multiple Sort- Sort score from highest to the lowest, sort the student age from the smallest to the largest
df.sort_values(by=['Score','Age'], ascending=[False, True]).reset_index(drop=True)
```

DATA ANALYSIS WITH PANDAS

```python
product_data1 = {
    'product_category': ['baby','book','book','beauty','baby'],
    'product_price': [19.99, 12.49, 22, 49, 16],
    'rating': [5,3,3,2,4]
}
df = pd.DataFrame(product_data1)
df
```

```python
#What is the average rating by product category?

avg_rating_summary = df.groupby('product_category')['rating'].mean().reset_index()

## renaming the rating column name to avg_rating
avg_rating_summary.rename(columns={'rating':'avg_rating'}, inplace=True)
avg_rating_summary


```

```python
## What is the average product price and total quantity of each product category
product_price_qty = {
    'order_id': ['4235','2342','1234','5325','1342'],
    'product_category': ['baby','beauty', 'baby','beauty', 'beauty'],
    'product': ['diaper','eye liner', 'diaper','eye shadow', 'face mask'],
    'product_price': [19.99, 12.49, 22, 49, 16],
    'quantity': [5,3,3,2,4]
}

df = pd.DataFrame(product_price_qty)

## this gives only the average product price. we need the total qty as well..
avg_product_price_total_qty = df.groupby('product_category')['product_price'].mean().reset_index()
# avg_product_price_total_qty

## for multiple criterias then use agg() function
avg_product_price_total_qty = df.groupby('product_category').agg({'product_price':'mean', 'quantity':'sum'}).reset_index()
print(df)

avg_product_price_total_qty.rename(columns={'product_price':'avg_product_price', 'quantity':'total_qty'}, inplace=True)
avg_product_price_total_qty
```
