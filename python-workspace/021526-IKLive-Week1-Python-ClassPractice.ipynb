```python
#contains duplicates problem
def contains_duplicates(nums):
    s1 = set(nums)
    if len(nums) != len(s1):
        return True
    return False

l1 = ["1","2","3","1"]
contains_duplicates(l1)
```

```python
#intersection two arrays
def intersection_two_arrays(nums1, nums2):
    set_nums1 = set(nums1)
    set_nums2 = set(nums2)
    return set_nums1.intersection(set_nums2)

print(intersection_two_arrays([1,2,3,5,6,1,2],[2,3,1]))
```

```python
#Two sum problem
def two_sums(nums, target):
    d1 = dict()
    for i in range(len(nums)):
        if nums[i] in d1.keys():
            continue
        else:
            d1[nums[i]]= i

    for i in range(len(nums)):
        complementary = target - nums[i]
        if complementary in d1:
            if d1[complementary] != i:
                return i, d1[complementary]
            else:
                continue

    return -1

print(two_sums([1,2,3,4,5,7],7))
```

Call by Reference and Call by Value Examnples

```python
#call by value
def my_pass_value(x):
    print (x, id(x))
    x=5
    print(x, id(x))


x=7
print(x, id(x))
my_pass_value(x)
print(x,id(x))
```

```python
#call by reference
def my_pass_ref(l2):
    print (l2, id(l2))
    l2.append(5)
    print(l2, id(l2))

l1 = [7]
print(l1, id(l1))
my_pass_ref(l1)
print(l1, id(l1))
```
