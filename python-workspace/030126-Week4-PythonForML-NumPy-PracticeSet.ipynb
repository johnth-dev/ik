```python
## Numpy-FF5: Write a Python program using NumPy to create a 10x10 matrix filled with random integers between 1 and 100. Find the minimum and maximum values of the matrix along with their indices.

#arr10 = np.random.randint(1,101,100).reshape(10,10)
# OR
arr10 = np.random.randint(1,101,size=(10,10))

print(arr10)
print(f"Min value of the matrix is {np.min(arr10)} and is at {np.argmin(arr10)} th position")
print(f"Max value of the matrix is {np.max(arr10)} and is at {np.argmax(arr10)} th position")

#argmin gives only the single index linear index. this solution doesn't work very well with 2D arrays/matrix

'''Better solution is subclass'''
min_val_idx = np.argmin(arr10)
max_val_idx = np.argmax(arr10)

##Unravel - provides the coordinates of value provides
min_idx = np.unravel_index(min_val_idx, arr10.shape)
max_idx = np.unravel_index(max_val_idx, arr10.shape)

print(min_idx, max_idx)




```

```python
## Numpy-FF8: Write a Python program using NumPy to create a 1D array of 20 elements with random integers between 1 and 100. Calculate the number of even and odd elements in the array and print the result.

#arr_x = np.random.randint(1,101,20).reshape(1,-1)
arr_x = np.random.randint(1,101,20)

print(arr_x)

even_arr = arr_x[arr_x %2 == 0]
print("Number of even numbers in the array:", len(even_arr))

odd_arr = arr_x[arr_x %2 != 1]
print("Number of odd numbers in the array:", len(odd_arr))

print(len(arr_x[arr_x % 2 == 0]))
print(len(arr_x[arr_x % 2 == 1]))

#Alternate solution
print(arr_x % 2)
print(np.count_nonzero(arr_x % 2 == 0)) #Even numbers
print(np.count_nonzero(arr_x % 2 == 1)) #Odd numbers

#Alternate solution
print(len(np.where(arr_x %2 == 0)[0]))
print(len(np.where(arr_x %2 == 1)[0]))
```
