```python
#IK Week2 - Python Practice Set
#Problem #7: *** Maximum Sum Subarray ***
#Details: Given a list of integers, find the sum of a non-empty subarray that has the largest sum. The elements needs to be contigous and single element is also considered as subarray.
#Input =[2, -6, 3, 4, -5]
#Output: 7

#numbers = [-7, -9, -3, -5]
# numbers = [-1, -3, -4, 1, 5, -2, -7]
numbers = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
#numbers = [-1,0,2]
sub_arr = []


# Solution 2: Instructor solution
# subarray can be one element. in case of all -ive elements, largest single -ive number will be answer
# If array is all positive, max sub array will be all +ive numbers and max sum will be the addition of all numbers
# subarray needs to be contiguous.
# Best O(n) time complexity approach will be Kadane's Algorithm. Sliding window approach, when the currsum becomes -ive, skip the previous number added and start the sum again from next element.
#numbers = [-7, -9, -3, -5]
n = len (numbers)
curr_sum = numbers[0]
max_sum = numbers[0]
sub_arr.append(numbers[0])

for i in range(1, n):
    if curr_sum < 0 :
        curr_sum = 0
    curr_sum = curr_sum + numbers[i]
    sub_arr.append(numbers[i])
    max_sum = max(curr_sum, max_sum)

print(max_sum)
#return max_sum
```
