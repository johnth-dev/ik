```python
def calculate_mmm(data, column):
  """
  Calculates mean, median, and mode of a numeric column in a DataFrame.

  Args:
    data: Pandas DataFrame.
    column: Name of the numeric column.

  Returns:
    A tuple containing mean, median, and mode.
  """

  #mean = # write your code here
  mean = data[column].mean()
  print(mean)
  median = data[column].median()
  mode = data[column].mode()
  #median = # write your code here
  #mode = # write your code here

  return mean, median, mode

mean_age, median_age, mode_age = calculate_mmm(titanic_data, 'age')

print("Mean age:", mean_age)
print("Median age:", median_age)
print("Mode age:", mode_age)
```

```python
# **Question 13**
def survival_stats(data):
  """
  Calculates survival percentages for males and females.

  Args:
    data: Pandas DataFrame containing Titanic data.

  Returns:
    None. Prints the calculated statistics.
  """

  # Calculate survival percentages
  # male_survived = # write your code here

  male_survived = data[ (data['sex'] == 'male') & (data['survived'] == 1 ) ].shape[0]

  print("male_survived:", male_survived)

  # female_survived = # write your code here

  female_survived = data[ (data['sex'] == 'female') & (data['survived'] == 1 ) ].shape[0]

  print("female_survived:", female_survived)
  #
  total_male = data[ data['sex'] == 'male'  ].shape[0]
  total_female =  data[ data['sex'] == 'female'  ].shape[0]

  print("total male:", total_male)
  print("total female:", total_female)

  percent_male_survived = (male_survived/total_male)*100.0
  percent_female_survived = (female_survived/total_female)*100.0

  print("Percentage of males who survived: {:.2f}%".format(percent_male_survived))
  print("Percentage of females who survived: {:.2f}%".format(percent_female_survived))

  print(data.groupby('sex')['survived'].mean()*100)

survival_stats(titanic_data)
```

```python
def most_common_embarkation(data):
    """
    Finds the embarkation port with the highest number of passengers.
    Args:
        data: pandas DataFrame containing the Titanic dataset.
    Returns:
        Most common embarkation port ('C','Q','S').
    """
    # write your code here
    # most_embarkation_port = data['embark_town'].mode()
    # return most_embarkation_port

    # Instructor answer
    embark_counts = data['embarked'].value_counts()
    return embark_counts.idxmax()

common_port = most_common_embarkation(titanic_data)
print("Most common embarkation port:", common_port)
```

```python
def impute_categorical_nan(data, column):
  """
  Imputes missing values in a categorical column with the most frequent value.

  Args:
    data: Pandas DataFrame.
    column: Name of the categorical column with missing values.

  Returns:
    Pandas DataFrame with imputed values.
  """

  # Find the most frequent value
  data.info()
  most_frequent_value = data[column].mode()[0]
  print(most_frequent_value)


  # Fill missing values with the most frequent value
  data[column] = data[column].fillna(most_frequent_value)# write your code here
  data.info()
  return data

df = impute_categorical_nan(titanic_data, 'embarked')

df.head()
```
